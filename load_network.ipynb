```python
##Whole python code:

##Read the file
import h5py
filename = "5gen-1mil.h5"

dataset = h5py.File(filename, 'r')


#Nodes indices: from to
grid = dataset['GEOM_GROUP']
indexes = list(grid['NODE_ARRAY'])
df = pd.DataFrame(indexes, columns=['node1', 'node2'])


#First vessels
vessel_names = ['vessel' + str(i) for i in range(len(df))]
df.insert(0, 'vessel', vessel_names)

#coordinates and radiuses
col_names = ['x0', 'y0', 'z0', 'x1', 'y1', 'z1', 'radius']
geom = list(grid['GEOM_ARRAY'])
df2 = pd.DataFrame(geom, columns=col_names)
df = pd.concat([df, df2], axis=1)


#read the dataset of flow group=> damaged and healthy
grid = dataset['FLOW_GROUP']
grid.keys()

d_flow = list(grid['DAMAGED_FLOW_ARRAY'])
df['damaged_flow']=d_flow

h_flow = list(grid['HEALTHY_FLOW_ARRAY'])
df['HEALTHY_FLOW_ARRAY']=h_flow


#save it in csv file
df.to_csv('Network_5g.csv', index=False)
```
